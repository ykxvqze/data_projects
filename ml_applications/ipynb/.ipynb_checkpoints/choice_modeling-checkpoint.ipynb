## Introduction

We'll perform choice-based conjoint analysis in order to study customer preferences for attributes of a given type of product. The product in this case is a tablet. A survey was conducted in which customers were asked to select one from among three alternatives presented to them. The attributes are of course varied across the alternatives. Every customer was shown 15 such choice sets (each with 3 alternatives), and therefore 15 selections were recorded.

The attributes of a tablet are: brand, screen size (in inches), HD storage, RAM capacity, battery life, and price (in dollars). We would like to understand the importance of these attributes for customers. 

## Data exploration

The first 3 rows show that customer with id = 1 selected the first alternative when presented with the first choice set or question. This is indicated by a "1" in the variable "choice". Thus this customer prefers an iPad tablet with a 7-inch screen, 32GB storage, 4GB ram, 7-hour battery life, and a price of \$499 over a Surface tablet with characteristics shown in the second row or a Kindle with characteristics shown in the third row.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#data exploration
path_to_file = './data/tablets.txt'
data = pd.read_csv(path_to_file)	
data.head(6)

#data.shape #137*15*3 x 10
```

## Multinomial logit model

We will build a conjoint multinomial model and train it over the data using the function **mnlogit()**.

Train the model using **mnlogit()** as shown below. Note that the "0" in the formula indicates that we do not wish to have an intercept included. 

```python
from statsmodels.formula.api import mnlogit

model = mnlogit('choice ~ 0 + brand + size + storage + ram + battery + price', data).fit()
model.summary();
```

## Parameter interpretation 

The attribute **summary()** reveals the coefficients associated with each attribute level. These coefficients can be compared (in sign and magnitude). Which brand has the highest (positive) coefficient? Also, does it make sense that the attribute "price" has a negative coefficient? Note that the attribute levels not appearing in the summary are reference levels and their coefficients are zero.     

## Prediction

Use **predict()** to predict the probability of selecting each alternative for the given choice sets. We can therefore now make customer choice predictions for new alternatives (and their competition) before we release these into the market. However, we would also like to estimate the market share based on these choice predictions. For that purpose, we may use the softmax function to squash the outputs into the range $[0,1]$ such that the sum equals 1. 

If $U_j$ represents the (soft) label returned by the model for selecting some alternative $j$, then with $n$ alternatives being compared, the probability of selecting alternative $j$ is $p_j = \dfrac{exp(U_j)}{exp(U_1)+exp(U_2)+...+exp(U_n)}$, $j \in \{1,2,...,n\}$ (softmax)

The probabilities sum up to 1: $(p_1+p_2+...+p_n=1)$

## Willingness-to-pay

From the parameter (coefficient) estimates of the model, we may compute how much a consumer would be willing to pay for a specific attribute level (such as 9-hour battery life). We just need to calculate the ratio of the coefficient for the specific attribute level to the coefficient of price (in absolute value). 

How much is the average consumer willing to pay in dollars to change tablet brand from Galaxy to iPad (all else fixed)? How much is a consumer willing to pay for an upgrade from 1GB to 4GB RAM? What about an upgrade from a 7-inch screen to a 9-inch screen?   

```python
#willingness-to-pay
model.params

galaxy_to_ipad = (model.params.loc['brand[Galaxy]'] - model.params.loc['brand[iPad]'])/model.params.loc['price']
galaxy_to_ipad[0]

r1gb_to_r4gb = (0 - model.params.loc['ram[T.r4gb]'])/model.params.loc['price']
r1gb_to_r4gb[0] 

sz7inch_to_sz9inch = (model.params.loc["size[T.sz7inch]"] - model.params.loc["size[T.sz9inch]"])/model.params.loc["price"]
sz7inch_to_sz9inch[0];
```
