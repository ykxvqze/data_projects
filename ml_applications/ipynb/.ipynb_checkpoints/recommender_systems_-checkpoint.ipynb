## Introduction

We'll build a simple recommender system. In particular, we will explore user-based and item-based collaborative filtering.

## Read-in data

Our data consists of real purchases made between 01.12.2010 and 09.12.2011 at a retail company that sells mostly online gift items. The data is provided in the file "invoices.csv". Inspect the data: how many observations and variables are there? 

```python
#read-in data
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

path_to_file = './data/invoices.csv'
data = pd.read_csv(path_to_file)
data.head()
#data.shape
```

## Data preparation

Inspect the "quantity" field, and check if there are any quantities that are not positive! Show the first few observations that have negative quantities. Negative values signify that the order was cancelled, and these transactions can be ignored for our purpose. Create a new data frame by ignoring these observations and keeping only those with `quantity > 0`. 

Next, inspect the field "customer_id". Are there any missing values? (hint: **isna()** or **isnull()**). How many are these? Remove any observation containing NaN values (e.g. by applying **dropna()** on the data frame). What is the size of the dataset now? 

```python
#data preparation
data.loc[data['quantity']<0].head(6) 
df = data.loc[data['quantity']>0]
df.loc[df['customer_id'].isna()] #or .isnull()
df['customer_id'].isna().sum() 
df = df.dropna(subset=['customer_id'])  #or df.loc[~df['customer_id'].isna()]
df.shape;
```

## Customer-Item matrix

Before building a recommender system, we need to set up a customer-item (binary) matrix where each row is a unique customer and each column is a unique item. A value of 1 in the matrix at location (i,j) indicates that customer (i) purchased item (j), whereas a 0 indicates that no purchase was made. 

Extract unique customer IDs and sort them in ascending order (hint: `unique()`; `sort()`). Do the same for the field "stock_code" which represents item IDs. Now, set up the user-item matrix where the rows correspond to the sorted customer IDs and the columns to the sorted stock codes. Try not to go beyond a single `for` loop when constructing the user-item matrix for this dataset. Alternatively, you may use pandas' **pivot_table()**. 

```python
#using: pivot_table()
user_item_matrix = pd.pivot_table(df, index='customer_id', columns='stock_code', values='quantity', aggfunc='sum', fill_value=0)
user_item_matrix.shape

df['stock_code'].nunique()
df['customer_id'].nunique()

user_item_matrix = user_item_matrix.applymap(lambda x: 1 if x > 0 else 0) #applymap() is over every element in data frame

#alternatively: 
'''
stock_codes = df['stock_code'].unique()
customer_ids = df['customer_id'].unique()

stock_codes = np.sort(stock_codes)
customer_ids = np.sort(customer_ids)

user_item_matrix = pd.DataFrame(0,columns=stock_codes,index=customer_ids)

df_items = df.groupby('customer_id').aggregate({'stock_code': 'unique'})

for idx,customer in enumerate(df_items.index):
    user_item_matrix.loc[customer][df_items.loc[customer][0]] = 1
    #user_item_matrix.loc[customer][df_items.loc[customer]['stock_code']] = 1 #same thing
'''
    
user_item_matrix.head(6);
```

## Collaborative filtering

### User-based collaborative filtering

In order to derive the user-user similarity matrix, we need to compute distances between users based on the cosine similarity measure which is simply the cosine of the angle between two vectors (users) i.e. rows from the user-item matrix. The cosine similarity measure can be computed efficiently with **cosine_similarity()** (module: **sklearn.metrics.pairwise**). Note that the output should be an $n \times n$ matrix, where $n$ is the number of unique customers. Make sure the index and column names of the user-user similarity matrix are set correctly (i.e. within a pandas data frame).

```python
#user-based collaborative filtering
from sklearn.metrics.pairwise import cosine_similarity

user_user_matrix = pd.DataFrame(cosine_similarity(user_item_matrix), columns=user_item_matrix.index, index=user_item_matrix.index)

#alternatively:
#user_user_matrix = cosine_similarity(user_item_matrix)
#user_user_matrix = pd.DataFrame(user_user_matrix, index=customer_id, columns=customer_id)

user_user_matrix.head();
```

### Making recommendations 

With the user-user similarity matrix in place, in addition to the user-item matrix, we are now ready to make recommendations. Suppose we wish to make recommendations for user A with customer ID: 13645. Find the IDs of the 10 nearest neighbors, i.e. most similar users, to customer A.   

Extract which items where bought by user A from the user-item matrix. Similarly, extract item IDs that were bought by the 10 most similar users to A (hint: one efficient way is to use **any()** over the customer-item submatrix corresponding to the 10 nearest neighbors). 

Identify items to recommend to A based on items bought by the 10 most similar users to A, but which A hasn't bought yet (hint: many ways are possible, however sets can be subtracted, so you may want to use sets for an efficient solution). Finally use the "description" field to obtain the descriptions of the recommended item IDs, and show both the recommended item IDs and their descriptions in the same data frame. Sort the data frame according to stock code in ascending order. 

```python
#making recommendations
customer_id = 13645
neighbors = user_user_matrix.loc[customer_id].sort_values(ascending=False).index[1:10+1]
neighbors = neighbors.tolist() #or .values to obtain an array instead

ind = np.where(user_item_matrix.loc[customer_id,:]==1)
items_bought_by_customer = user_item_matrix.columns[ind].tolist()

#or
#items_bought_by_customer = user_item_matrix.columns[user_item_matrix.loc[customer_id,:]==1].tolist()
#or
#items_bought_by_customer = user_item_matrix.loc[customer_id, user_item_matrix.loc[customer_id,:]==1].index.tolist()
#or 
#items_bought_by_customer = df_items.loc[customer_id]['stock_code'].tolist()
 
ind = user_item_matrix.loc[neighbors].any(axis=0)
items_bought_by_neighbors = user_item_matrix.columns[ind].tolist()

items_to_recommend = list(set(items_bought_by_neighbors) - set(items_bought_by_customer)) #sets can be subtracted, not lists

#alternatively: list comprehension
#items_to_recommend = [x for x in items_bought_by_neighbors if x not in items_bought_by_customer]

#alternatively: using .remove() from list
#items_to_recommend = items_bought_by_neighbors
#for item in items_bought_by_customer:
#    if item in items_bought_by_neighbors: items_bought_by_neighbors.remove(item)

items_to_recommend_descriptions = df.loc[df['stock_code'].isin(items_to_recommend), ['stock_code','description']].drop_duplicates().set_index('stock_code')

#alternatively: (though slower), note: results will be slight different since the duplicates above are row duplicates in one shot overall
#items_to_recommend_descriptions = [df.loc[df.stock_code == code][['stock_code','description']].iloc[0] for code in items_bought_by_neighbors] #choose first item
#items_to_recommend_descriptions = pd.DataFrame({'stock code' : items_bought_by_neighbors, 'description': items_to_recommend_descriptions})

items_to_recommend_descriptions.sort_values(by=['stock_code'], ascending = True); 
```

### Item-based collaborative filtering

Obtain the item-item similarity matrix again using **cosine_similarity()**; note that the input to the function in this case should be the transpose of the user-item matrix we created previously, and the output should be an $m \times m$ matrix, where $m$ is the number of unique item IDs (stock codes). Make sure to set the index and column names. Using this similarity matrix, identify the stock codes of the 10 most similar items to the item with stock code 71053. Finally extract the descriptions of the item with stock code 71053 and its 10 most similar items, and show these along with their stock codes in the same data frame. 

```python
#item-based collaborative filtering
item_item_matrix = pd.DataFrame(cosine_similarity(user_item_matrix.T), columns=user_item_matrix.columns, index=user_item_matrix.columns)
item_item_matrix.head()

stock_id = '71053' #since stored as strings, where customer_id were interpreted as floats in pandas 
similar_items = item_item_matrix[stock_id].sort_values(ascending=False).index[0:10+1] #keep the 0th
#neighbors = neighbors.tolist() #or .values to obtain an array instead

items_to_recommend_descriptions = df.loc[df['stock_code'].isin(similar_items), ['stock_code','description']].drop_duplicates().set_index('stock_code')
items_to_recommend_descriptions.sort_values(by=['stock_code'], ascending = True);
```
