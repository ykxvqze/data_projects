## Introduction

We'll analyze customer conversion rate as a marketing KPI.

The dataset is gathered from a bank marketing campaign based on phone calls to bank clients in the aim of subscribing clients to a term deposit. Conversion in this context is based on whether the client subscribed ("yes") or not ("no") to a term deposit.

## Data exploration

The data is provided in the file "conversion.txt". Read in the file and inspect the data. The following is a descriptive summary of the attributes and target variable.  

```
Input variables:
# bank client data:
1 - age (numeric)
2 - job : type of job 
3 - marital : marital status 
4 - education 
5 - default: has credit in default? 
6 - housing: has housing loan? 
7 - loan: has personal loan? 
# related with the last contact of the current campaign:
8 - contact: contact communication type
9 - month: last contact month of year 
10 - day_of_week: last contact day of the week 
11 - duration: last contact duration, in seconds. Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no'). Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.
# other attributes:
12 - campaign: number of contacts performed during this campaign and for this client 
13 - pdays: number of days that passed by after the client was last contacted from a previous campaign (999 means client was not previously contacted)
14 - previous: number of contacts performed before this campaign and for this client 
15 - poutcome: outcome of the previous marketing campaign 
# social and economic context attributes
16 - emp_var_rate: employment variation rate - quarterly indicator 
17 - cons_price_idx: consumer price index - monthly indicator
18 - cons_conf_idx: consumer confidence index - monthly indicator 
19 - euribor3m: euribor 3 month rate - daily indicator 
20 - nr.employed: number of employees - quarterly indicator 
#Output variable (desired target):
21 - y - has the client subscribed a term deposit?
```

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#data exploration
path_to_file = './data/conversion.txt'
data = pd.read_csv(path_to_file)
data.head()
data.shape;
````

## Conversion rates

Transform the conversion variable into a binary numeric variable where 1 represents "yes" and 0 represents "no" (hint: **apply()** an anonymous `lambda`, or **replace()**, etc.). Compute the total number of conversions out of all responses, and indicate the overall conversion rate as a percentage. 

````python
#conversion rates
data['y'] = data['y'].apply(lambda x: 1 if x=='yes' else 0)

#alternatively:
#data['y'].replace({'yes':1, 'no':0}, inplace=True)

#alternatively:
#data['y'].replace(to_replace=['yes', 'no'], value=[1, 0], inplace=True)

data['y'].sum()/data['y'].count() * 100; #or data['y'].sum()/data.shape[0] * 100
````

### By number of contacts

We will analyze conversion as a function of the number of contacts conducted by the bank to a client. Create a data frame consisting of the number of contacts (hint: `campaign`) as index, and three columns: the count for each number of contacts in the data, the total number of conversions (hint: pandas' **groupby()**; **aggregate()**), and the conversion rate (as percentage) derived from the previous two values. Make sure you assign meaningful names to the columns. Show the first few rows and plot the conversion rates as a function of the number of contacts for the first 10 elements (i.e. number of contacts ranging from 1 to 10).              

````python
#by number of contacts
df = data.groupby(['campaign']).aggregate({'campaign': 'count', 'y': 'sum'})
df.rename(columns={'campaign': 'count', 'y': 'n_conversions'}, inplace=True)
df['conversion_rate'] = df['n_conversions']/df['count'] * 100
#df.index.name = '...' #if you want to change index name


plt.plot(df.index.values[:10], df['conversion_rate'].values[:10], c='b') #or df.loc[:10,'conversion_rate'] 
plt.scatter(df.index.values[:10], df['conversion_rate'].values[:10], c='b')
plt.xlabel('number of contacts')
plt.ylabel('conversion rate (%)')
plt.show()
````

This is due to the fact that the sample size (number of clients) becomes small for high ages (you may check how the count changes with age in your previous data frame). Therefore, to reduce noise, we will set up age groups and create a bar plot showing the conversion rate for each group. ## Duration of last contact 

Compute the duration (in hours) of the last contact with the bank for each client (hint: `duration` column). Note that the duration in the dataset is originally provided in seconds (as mentioned in the attribute descriptions above). Create two horizontal box plots in the same figure summarizing the spread of duration values for the case of conversions and non-conversions.   

````python
#duration of last contact
h = plt.boxplot([data.loc[data['y']==0,'duration']/3600, data.loc[data['y']==1,'duration']/3600], labels=[0, 1], patch_artist=True, vert=False)
plt.xlabel('duration (in hours)')
plt.ylabel('conversion status')
plt.title('last contact duration')
#plt.yticks(ticks=[1,2], labels=[0,1]) #no need for this since labels specified in boxplot()
plt.setp(h["boxes"][0], color="red")
plt.setp(h['boxes'][1], color="blue")
plt.show() 
````

## Logistic regression

Construct a logistic regression model for predicting conversion (0 or 1) from the attributes: "age", "campaign" (number of contacts), "previous", "marital", "job", and "housing". Transform categorical variables first by one-hot-encoding (e.g. **pd.get_dummies()**). You may use **LogisticRegression()** from **sklearn.linear_model**. Inspect the model coefficients (attribute `coef_`). Which attributes are positively/negatively correlated with conversion. Sort the coefficients along with their names.

````python
#logistic regression
df_categorical = pd.get_dummies(data[['marital','job','housing']]) #can add option drop_first=True
df = pd.concat([df_categorical, data[['age','campaign','previous']]], axis=1)

X = df.copy()
y = data['y']

from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')

model = LogisticRegression(solver='lbfgs').fit(X, y)
y_pred = model.predict(X)

model.score(X, y) #accuracy
model.coef_
model.intercept_

df = pd.DataFrame(model.coef_[0], index=X.columns.tolist(), columns=['coefficients'])

df.loc[df['coefficients']>0]
df.loc[df['coefficients']<0]

df.sort_values(by=['coefficients'], ascending=True);
````
